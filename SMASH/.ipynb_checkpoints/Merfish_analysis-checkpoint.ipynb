# Analysis of Mouse hypothalamus data collected using MERFISH

## Data description

The mouse hypothalamus data has 161 genes and 5665 cells. 156 genes are preselected
markers for different cell types, whereas the other five are control genes. Those 156 genes can be expected to be highly variable.

## Loading the package and the data

We import the main function SMASH and read the gene expression data (as, 'Gene') and the spatial co-ordinates (as, 'Cords'). 

The gene expression data and the spatial co-ordinates must be provided as pandas dataframes. For the former, rows must correspond to n cells/spots and columns must correspond to p individual genes (i.e, the dimension is n x p). The column names of this dataframe must be set to the gene names. For the second dataframe, rows must correspond to the same cells/spots (in exactly same order) as the first dataframe, and its two columns should correspond to x and y co-ordinates (i.e, the dimension is n x 2).  

```python
from SMASH import SMASH, Expression_plot
import pickle
import pandas as pd
from statsmodels.stats.multitest import multipletests

path = "/Users/sealso/Documents/GitHub/SMASH"
with open(path + "/Data/Mfish.pickle", "rb") as fp:  
 df = pickle.load(fp) 

Gene = df[0]; Cords = df[1];
pd.set_option('display.float_format', '{:.2g}'.format)
```

## Previewing the loaded data

To showcase how the dataframes should look, we preview the first few rows of both the dataframes. Notice that the row-names of the two dataframes are matching which implies that the cells/spots are in the same order in both the dataframes.

```python
Gene.head()
```

```python
Cords = Cords.set_axis(["X", "Y"], axis = 1) 
Cords
```

## Running SMASH

Next, we run the function SMASH with these two dataframes as arguments.

Some additional arguments that can be passed are, len_l, mean_only, and kernel_covariance.

1. len_l, which is equal to 10 by default, specifies the number of values to use as the lengthscale (or, period) parameter. 
2. mean_only, which is kept to be "FALSE", is a TRUE/FALSE string specifying either just a mean-based test (SPARK-X) to be performed or not. 
3. kernel_covariance, which is kept to be "All", is a string argument specifying if both the Gaussian and Cosine kernel covariance matrices to be used in the covariance-based tests. It can be changed to either "Gaussian" or "Cosine" if only one particular type is of interest. Note that if mean_only = "TRUE", this parameter would not be used as the test would not remain covariance-based anymore.

```python
Run_method = SMASH(Gene, Cords)
```

## Exploring the results

The above function returns three dataframes with three types of p-values, one of which corresponds to SMASH. This dataframe has two columns with the first column having the gene names and the second column having 
the p-values corresponding to the actual SMASH test. 
 
The other two dataframes correspond to the SPARK-X method i.e., just a mean-based test, and to an apparoximate version of the SpatialDE method. Further details can be found in the supplementary material of the main paper. 

```python
Run_method.keys() # Checking the content of the obtained result
```

## Extracting SMASH p-values

Next, we extract the dataframe with the SMASH p-values and view first its first few rows. 

```python
SMASH_pvals = Run_method['SMASH']
SMASH_pvals.head() 
```

## Benjamini/Yukateli p-value correction for FWER control

We adjust the p-values using Benjamini/Yukateli procedure to control the family wise error rate (FWER) at 0.01.

```python
SMASH_pvals['Adjusted p-val'] = multipletests(SMASH_pvals['p-val'], alpha=0.01, method='fdr_by')[1]
SMASH_pvals.head()
```

## Total number of detected SVGs 

We check how many genes are detcted to be SVGs at FWER = 0.01, and extract their names. Note that 'pandas' round off very small p-values to 0 which is why all the p-values below appear to be 0. 

```python
SVGS = SMASH_pvals[SMASH_pvals['Adjusted p-val'] < 0.01]
print(SVGS)
print({'# detected SVGs ': SVGS.shape[0]})
```

## Visualize the spatial expression of particular SVGs

We display the relative expression of an SVG named "Npy1r" in different cells below. 


```python
Gene_name = 'Npy1r'
Data = pd.concat([Cords, Gene[Gene_name]], axis = 1)
adj_pval = SMASH_pvals[SMASH_pvals['Gene'] == Gene_name]['Adjusted p-val']
print({Gene_name + ' p-value ': adj_pval.values}) # Printing the adjusted p-value of the SVG
Expression_plot(Data, Gene_name, s = 0.5, cmap = 'viridis_r') # Displaying the expression plot of the SVG
```

Lets visulaize two more SVGs, "Rxfp1" and "Gad1".

```python
Gene_name = 'Rxfp1'
Data = pd.concat([Cords, Gene[Gene_name]], axis = 1)
adj_pval = SMASH_pvals[SMASH_pvals['Gene'] == Gene_name]['Adjusted p-val']
print({Gene_name + ' p-value ': adj_pval.values}) # Printing the adjusted p-value of the SVG
Expression_plot(Data, Gene_name, s = 0.5, cmap = 'viridis_r') # Displaying the expression plot of the SVG


Gene_name = 'Gad1'
Data = pd.concat([Cords, Gene[Gene_name]], axis = 1)
adj_pval = SMASH_pvals[SMASH_pvals['Gene'] == Gene_name]['Adjusted p-val']
print({Gene_name + ' p-value ': adj_pval.values}) # Printing the adjusted p-value of the SVG
Expression_plot(Data, Gene_name, s = 0.5, cmap = 'viridis_r') # Displaying the expression plot of the SVG
```

"Expression_plot" is a very basic visulaziation function. For more sophisticated plots, additional R functions have been provided that we have used in the main paper.
